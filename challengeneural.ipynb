##  Calcul du mod√®le 

```python
data=pd.read_csv('dataNormalized.csv',index_col=0)
train,test=train_test_split(data,test_size=0.1,stratify=data['y'])
```

**Formule empirique pour approximer le nombre de neuronne par couche:**

$\begin{align} 
    N_h &=\frac{N_s}{(\alpha * (N_i + N_s))}\\
    &= \frac{3 166 465}{(2:10)*37}
    \end{align}$

```python
testp=[]
for a in range(2000,4000,500):
    testp.append((a))
    for b in range(2000,4000,500):
        testp.append((a,b))
testp
```

```python

paramGrid_lat = {
        'hidden_layer_sizes':testp,
       #'batch_size' : [80,85,90,95,100], best=95      
    }


estimator = MLPClassifier( random_state=1,batch_size=95)

gsc = DSC.GridSearchCV( estimator, paramGrid_lat, iid=False,cv=2,n_jobs=-1)
```

```python
#with joblib.parallel_backend("dask",scatter=[train.drop('y',axis=1),train.y]):
#   print('start')
gsc.fit(train.drop('y',axis=1),train.y)
```

## Sauvegarde des r√©sultats et affichage

```python
result = pd.DataFrame(gsc.cv_results_).sort_values('rank_test_score')
result.to_csv('NN.csv',index=None)
result
```

```python
filename = 'NN_final.sav'
joblib.dump( gsc, filename)
```
